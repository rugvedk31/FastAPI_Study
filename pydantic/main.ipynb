```python
# Without using the pydantic how the invalid data can pass using dataclass
from dataclasses import dataclass

@dataclass
class person:
    name: str
    age: int
    city: str
    
person1 = person(name="Alex", age=20, city=12) # This will not raise any error because dataclass does not perform data validation
print(person1)
```

```python
from pydantic import BaseModel
```

```python
#How data validation works in the pydantic 
```

```python
class person(BaseModel):
    name:str
    age:int
    city:str
    
person1 = person(name="Alex",age=20, city=12) # This will raise an error because pydantic performs data validation and city should be a string not an integer
print(person1)
```

```python
# Valid Data
class person(BaseModel):
    name:str
    age:int
    city:str
    
person1 = person(name="Alex",age=20, city="New York")
print(person1)
```
